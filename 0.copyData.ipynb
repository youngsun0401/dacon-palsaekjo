파일이 넘 크니까 앞 부분만 잘라서 저장.

```python
import pandas as pd

OUTPUT_SIZE = 2000

from mymodules.paths import Paths
PATH_IN = Paths('open_v2')
PATH_OUT = Paths('.')
```

## train 데이터 잘라넣기

```python
# train 데이터 읽기
train_A    = pd.read_csv(PATH_IN.train_A,    index_col='Test_id')
train_B    = pd.read_csv(PATH_IN.train_B,    index_col='Test_id')
train_meta = pd.read_csv(PATH_IN.train_meta, index_col='Test_id')

print("train_A:", train_A.shape)
print("train_B:", train_B.shape)
print("train_meta:", train_meta.shape)
```

```python
# train: 일정크기 앞부분
train_A = train_A.head(OUTPUT_SIZE)
train_B = train_B.head(OUTPUT_SIZE)
train_A.to_csv(PATH_OUT.train_A)
train_B.to_csv(PATH_OUT.train_B)
```

```python
# meta(정답): train에 해당하는 행들만
train_meta = train_meta.loc[train_meta.index.intersection(train_A.index.union(train_B.index))]

train_meta.to_csv(PATH_OUT.train_meta)
```

```python
# 확인용 출력
meta_A = train_meta[train_meta["Test"] == "A"].copy()
meta_B = train_meta[train_meta["Test"] == "B"].copy()

print(f"A set after alignment: train_A={len(train_A)}, meta_A={len(meta_A)}")
print(f"B set after alignment: train_B={len(train_B)}, meta_B={len(meta_B)}")
```

## test 데이터는 자르지는 않고 정렬

```python
test_A    = pd.read_csv(PATH_IN.test_A,    index_col='Test_id')
test_B    = pd.read_csv(PATH_IN.test_B,    index_col='Test_id')
test_meta = pd.read_csv(PATH_IN.test_meta, index_col='Test_id')

test_A    .sort_index().to_csv(PATH_OUT.test_A)
test_B    .sort_index().to_csv(PATH_OUT.test_B)
test_meta .sort_index().to_csv(PATH_OUT.test_meta)
```
